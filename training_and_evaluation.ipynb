### Erstes Training
Im nächsten Schritt trainieren wir zwei verschiedene Modelle, um sie zu untersuchen und zu vergleichen. Als Vorbereitung für das Training teilen wir die Daten in Trainings- und Testdaten auf.

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest= train_test_split(X, Y, test_size=0.15, random_state=1337)
display(xtrain.head())
```

Da wir mehrere Outputs haben und die Output beliebige Werte zwischen 0-100 annehmen können, handelt es sich hier um ein Multi-Target-Regressions-Problem.

Die Wahl der Regressoren fiel hier auf [GradientBoostingRegressor](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.GradientBoostingRegressor.html) und [RandomForestRegressor](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestRegressor.html) der sklearn Bibliothek. Diese bieten mit Hilfe der [MultiOutputRegressor](https://scikit-learn.org/stable/modules/generated/sklearn.multioutput.MultiOutputRegressor.html?highlight=multioutputregressor#sklearn.multioutput.MultiOutputRegressor) Klasse einen intuitiven Weg Multi-Target-Regressions-Probleme zu bearbeiten.

```python
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
from sklearn.multioutput import MultiOutputRegressor
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import r2_score as r2
from sklearn.metrics import explained_variance_score as evs

def createResultDataframe(testY, predictY):
    '''Returns a DataFrame with the same labels and dimensions as ytest'''
    return pd.DataFrame(index=testY.index,columns=list(testY),data=predictY)

modelBoostSimple = MultiOutputRegressor(estimator=GradientBoostingRegressor(random_state=1337))
predictYBoost = createResultDataframe(ytest, modelBoostSimple.fit(xtrain, ytrain).predict(xtest))

modelForestSimple = MultiOutputRegressor(estimator=RandomForestRegressor(random_state=1337))
predictYForest = createResultDataframe(ytest, modelForestSimple.fit(xtrain, ytrain).predict(xtest))

#Metriken mean_squared_error, r2_score und explained_variance_score andwenden
resultData = [[mse(ytest,output),r2(ytest,output), evs(ytest,output)] for output in [predictYBoost, predictYForest]]
resultDataFrame = pd.DataFrame(index=["GradientBoosting","RandomForest"]
                               ,columns=["MeanSquared","r2","explainedVariance"]
                               ,data=resultData)
print("Auswertung der ersten Modelle:")
display(resultDataFrame)
```

Die ersten beiden Modelle liefern keine besondern guten Vorhersagen, wie man in der obrigen Tabelle ablesen kann. Ein r^2 Score von ca 0 bedeutet, dass die Modelle nicht besser als konstante Modelle sind.

Es kann verschiedene Gründe geben, warum die Modelle schlecht sind. Einer ist, dass die Modelle ein niedriges lokales Maximum erzielt haben. Ein anderer wäre, ungeschickte Trainings- und Testdaten gewählt zu haben.

Dies wollen wir untersuchen.

### Wiederholungen des Trainings
Hier testen wir die Varianz der Modelle auf den gleichen Trainingsdaten bei mehrfachen Trainingsvorgängen.

Zunächst trainern wir die Modelle und berechnen die Metrik-Scores für jede Instanz.

**Achtung**: die mehrfachen Trainingsvorgänge benötigen Zeit. Die Anzahl der Wiederholungen kann über die Variable `repeat` beliebig erhöht werden.

```python
from numpy.random import randint, seed, RandomState
import numpy as np
from copy import deepcopy
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import r2_score as r2
from sklearn.metrics import explained_variance_score as evs

repeat = 10
seed = 1337

def repeat_training(x_train, y_train, regressor, repeat, seed):
    random_state = RandomState(seed)
    seeds = random_state.randint(10000000,size=repeat)
    for i in range(repeat):
        temp_reg = deepcopy(regressor)
        temp_reg.set_params(estimator__random_state=seeds[i])
        yield temp_reg.fit(x_train, y_train)    


def scores(testY, predictY):
    return [mse(testY,predictY),r2(testY,predictY), evs(testY,predictY)]
    

modelBoost = MultiOutputRegressor(estimator=GradientBoostingRegressor(random_state=seed))
modelForest = MultiOutputRegressor(estimator=RandomForestRegressor(random_state=seed))

resultBoost   = np.asarray([scores(ytest,output.predict(xtest)) 
                            for output 
                            in repeat_training(xtrain,
                                               ytrain,
                                               modelBoost,
                                               repeat,
                                               seed)])

resultForest  = np.asarray([scores(ytest,output.predict(xtest)) 
                            for output
                            in repeat_training(xtrain,
                                               ytrain,
                                               modelForest,
                                               repeat,
                                               seed)])
```

```python
import matplotlib.pyplot as plt
data = [np.asarray([resultBoost.T[i],resultForest.T[i]]).T for i in range(3)]
fig, ax = plt.subplots(3,sharex=True,figsize=(10,10))
_ = ax[0].boxplot(data[0],labels=["",""])
_ = ax[0].set_ylabel("MSE")
_ = ax[0].set_title("Mean Squared Error Box Plot")
_ = ax[1].boxplot(data[1],labels=["",""])
_ = ax[1].set_ylabel("r^2")
_ = ax[1].set_title("r^2 Score Box Plot")
_ = ax[2].boxplot(data[2],labels=["Gradient Boosting","Random Forest"])
_ = ax[2].set_ylabel("explained Variance")
_ = ax[2].set_title("Explained Variance Box Plot")
plt.show()
fig.savefig("./latex/figure/localminimum.pdf", 
            format='pdf', 
            bbox_inches = 'tight',
            pad_inches = 0
           )
```

Im obigen Plot sieht man den Vergleich zwischen Gradient Boost und Random Forest. Eine kurze Beschreibung zu den verwendeten Metriken befindet sich in unserem Report. Eine wichtige Beobachtung sollte sein, dass beide Modelle einen r^2 und explained Variance Wert von ungefähr 0 haben. Damit sind die Modelle insgesamt nicht besser als ein konstantes Modell. Insgesamt schneidet Gradient Boost etwas schlechter ab.

### Crossvalidation
Im nächsten Schritt untersuchen wir die Abhänigkeit der Modelle von der Wahl der Trainingsdaten. Dazu führen wir eine Wiederholende-KFold-Crossvalidation druch. Die Klasse ``RepeatedKFold`` teilt die Daten in zufällige gleichgroße Teilmengen (Splits) und führt Validation so durch, dass jeder Split einmal die Testmenge ist, während die restlichen die Trainingsmenge darstellen. Dieser Prozess wird wiederholt mit der Garantie, dass jedesmal andere Splits genommen werden. Dadurch ist sichergestellt, dass jeder Datenpunkt gleich oft Teil der Trainingsmenge und ebenso gleich oft Teil der Testmenge Menge ist. Für die Crossvalidierung benutzen wir die nützliche SKLearn-Methode `cross_validate`.

```python
from copy import deepcopy
from sklearn.model_selection import train_test_split, cross_validate, RepeatedKFold
from sklearn.metrics import mean_squared_error as mse, r2_score as r2, explained_variance_score as evs, make_scorer
from numpy.random import RandomState
n_splits = 5
n_repeats = 10

boosting = MultiOutputRegressor(estimator=GradientBoostingRegressor())
forest = MultiOutputRegressor(estimator=RandomForestRegressor())
scoring = {'mse':make_scorer(mse),'r2':make_scorer(r2),'evs':make_scorer(evs)}
rkf = RepeatedKFold(n_splits=n_splits, n_repeats=n_repeats,random_state=1337)
data_cross_boosting = cross_validate(boosting, X, Y,scoring=scoring,cv=rkf)
data_cross_forest = cross_validate(forest, X, Y,scoring=scoring,cv=rkf)
df_cross_boosting = pd.DataFrame.from_dict(data_cross_boosting)
df_cross_forest = pd.DataFrame.from_dict(data_cross_forest)

```

```python
data = [np.asarray([df_cross_boosting[i],
                  df_cross_forest[i]]).T for i in ['test_mse','test_r2','test_evs']]
fig, ax = plt.subplots(3,sharex=True,figsize=(10,10))
_ = ax[0].boxplot(data[0],labels=["",""])
_ = ax[0].set_ylabel("MSE")
_ = ax[0].set_title("Mean Squared Error Box Plot")
_ = ax[1].boxplot(data[1],labels=["",""])
_ = ax[1].set_ylabel("r^2")
_ = ax[1].set_title("r^2 Score Box Plot")
_ = ax[2].boxplot(data[2],labels=["Gradient Boosting","Random Forest"])
_ = ax[2].set_ylabel("explained Variance")
_ = ax[2].set_title("Explained Variance Box Plot")
plt.show()
```

Auch die Crossvalidation zeigt, dass Gradient Boosting insgesamt etwas schlechter abschneidet als Random Forest. Die Höhe der Boxplot zeigt an, dass durchaus eine gewisse Abhängigkeit zwischen gewählten Testdaten und Qualität der Modelle besteht. Da aber insbesondere der r^2  Wert bei allen Modellen ungefähr Null (oder kleiner) ist, können wir sicher sagen, dass die Modelle keine nützlichen Vorhersagen liefern und das unabhängig von den gewählten Trainingsdaten.

## Analyse der einzelnen Parteien
Nachdem wir die Modelle untersucht haben und festgestellt haben, dass beide Modelle nicht wirklich gute Vorhersagen liefern, wollen wir untersuchen ob die Modelle nicht doch für bestimmte Parteien gute Vorhersagen liefern können.
Zunächst untersuchen  wir nun die Vorhersagen für die einzelnen Parteien der ersten beiden Modelle die wir trainiert hatten und vergleichen diese mit der Ground Truth. Wir beschränken uns dabei auf die für uns interessanten Parteien CDU, SPD, GRÜNE, DIE LINKE, AfD und FDP.


```python
df_ypredGradientBoosting = pd.DataFrame(modelBoostSimple.predict(xtest) ,index=ytest.index,columns=list(ytest))
df_ypredRandomForestRegressor = pd.DataFrame(modelForestSimple.predict(xtest),index=ytest.index,columns=list(ytest))
display(df_ypredGradientBoosting.head())
display(df_ypredRandomForestRegressor.head())
```

```python
import numpy as np
#
import matplotlib.pyplot as plt
def plotAll(dfs,columnNames,parties=None,shape=None):
    columns = list(ytest) if parties == None else parties
    if shape == None:
        shape=(len(columns),1)
    columns = np.asarray(columns).reshape(shape)
    fig, axes = plt.subplots(nrows=shape[0],
                             ncols=shape[1], 
                             sharex=True, 
                             sharey=True, 
                             figsize=(10,10)
                            )

    for i, stadtteil in np.ndenumerate(columns):
        df_stadtteil = pd.DataFrame(index=ytest.index)
        df_stadtteil['ground truth'] = ytest[stadtteil]
        for j, df in enumerate(dfs):
            df_stadtteil[columnNames[j]] = df[stadtteil]
        df_stadtteil.plot.bar(ax=axes[i],legend=False)
        axes[i].set_ylabel(stadtteil)

    for i, _ in np.ndenumerate(columns):
        axes[i].yaxis.label.set_visible(True)
        axes[i].tick_params(
            axis = 'x',    
            which = 'both',
            bottom = (i[0]==shape[0]-1), 
            top = (i[0]==0),   
            labelbottom = (i[0]==shape[0]-1),
            labeltop = (i[0]==0),
            rotation=70
        )
    axes[0][shape[1]-1].legend(loc="center",bbox_to_anchor=(0.4,0.3))   
        
big_parties = ['Anteil CDU','Anteil SPD','Anteil GRÜNE','Anteil DIE LINKE','Anteil AfD','Anteil FDP']
plotAll([df_ypredGradientBoosting, df_ypredRandomForestRegressor]
        ,['GradientBoosting','RandomForestRegressor'],parties=big_parties,shape=(3,2))
```

Wir können sehen, dass die Vorhersagen der Modelle nicht willkürlich schlechte Ergebnisse liefern. Chancen bestehen, dass bestimmte Parteien gute Vorhersagen getroffen werden.

### Crossvalidation Parteien
Die letzte Untersuchung die wir durchführen wollen ist eine Wiederholung der Crossvalidation. Diesmal untersuchen wir jedoch die Vorhersagen für die einzelnen Parteien, anstatt die Gesamtvorhersage zu beurteilen. Leider bietet `cross_validate` nicht die Möglichkeit automatisch die Vorhersagen für die einzelnen Parteien auszuwerten, sodass wir dies implementieren mussten. Wir benutzen debei die gleichen RepeatedKFold-Splits wie in der vorherigen Crossvalidierung.
**Achtung**: Auch diese Durchführung braucht Zeit

```python
myX = X.to_numpy()
myY = Y.to_numpy()

dataSplit = list()
for train_index, test_index in rkf.split(X):
    X_train, X_test = myX[train_index], myX[test_index]
    Y_train, Y_test = myY[train_index], myY[test_index]
    myBoost = deepcopy(boosting).fit(X_train,Y_train)
    myForest = deepcopy(forest).fit(X_train,Y_train)
    dataSplit.append((myBoost,myForest,X_test,Y_test))
```

```python
from sklearn.metrics import mean_squared_error as mse,r2_score as r2, explained_variance_score as evs, make_scorer

scores_forest_r2 = list()
scores_forest_mse = list()
scores_forest_evs = list()

scores_boost_r2 = list()
scores_boost_mse = list()
scores_boost_evs = list()

for myBoost,myForest, X_test, Y_test in dataSplit:
    Y_boost = myBoost.predict(X_test)
    Y_forest = myForest.predict(X_test)
    scorer = lambda  metrik, Y_pred: metrik(Y_test, Y_pred, multioutput='raw_values')
    
    scores_forest_r2.append(scorer(r2,Y_forest))
    scores_forest_mse.append(scorer(mse,Y_forest))
    scores_forest_evs.append(scorer(evs,Y_forest))

    scores_boost_r2.append(scorer(r2,Y_boost))
    scores_boost_mse.append(scorer(mse,Y_boost))
    scores_boost_evs.append(scorer(evs,Y_boost))
```

```python
from sklearn.metrics import mean_squared_error as mse,r2_score as r2, explained_variance_score as evs, make_scorer

makepd = lambda data : pd.DataFrame(data=data,columns=list(Y))[big_parties]   
    
fig, axes = plt.subplots(3,2,figsize=(10,10),sharex=True,sharey='row')

make_boxplot = lambda data,i,j: makepd(data).boxplot(ax=axes[i][j])

make_boxplot(scores_forest_r2,0,0)
make_boxplot(scores_forest_mse,1,0)
make_boxplot(scores_forest_evs,2,0)
make_boxplot(scores_boost_r2,0,1)
make_boxplot(scores_boost_mse,1,1)
make_boxplot(scores_boost_evs,2,1)



axes[0][0].set_title('RandomForest')
axes[0][1].set_title('GradientBoost')
axes[0][0].set_ylabel('R^2 Score')
axes[1][0].set_ylabel('Mean Squared Error')
axes[2][0].set_ylabel('Explained Variance')
axes[2][0].tick_params(
    axis = 'x',  
    rotation=20
)
axes[2][1].tick_params(
    axis = 'x',  
    rotation=20
)

plt.show()
```

In diesem Boxplot können wir sehen, dass für die Parteien SPD, AFD und FDP deutlich bessere Ergebnisse erzielt werden konnten als für die restliche Parteien. Inbsondere der R^2 Wert ist im Schnitt deutlich über 0. Es gibt jedoch Ausreißer, sodass hier starke Abhängigkeit zu den gewählten Trainingsdaten zu vermuten ist. Für SPD und FDP ist der R2-Score sogar im Schnitt ungefährr 0.5. Aus den Daten der Einkommen lassen sich also nützliche Vorhersagen zum Wahlverhalten treffen. Dies unterstützt unsere Vermutung, dass es Abhängikeiten zwischen Einkommen und Wahlverhalten gibt.
