```python
import matplotlib.pyplot as plt
def plotAll(dfs,columnNames):
    fig, axes = plt.subplots(nrows=len(list(ytest)),
                             ncols=len(dfs), 
                             sharex=True, 
                             sharey='row', 
                             figsize=(10,50)
                            )
                            
    for i,stadtteil in enumerate(list(ytest)):
        for j, name in enumerate(columnNames):
            axes[0,j].set_title(name)
        for j, df in enumerate(dfs):
            df_stadtteil = pd.DataFrame(index=ytest.index)
            df_stadtteil['predicted'] = df[stadtteil]
            df_stadtteil['ground truth'] = ytest[stadtteil]
            df_stadtteil.plot.bar(ax=axes[i,j], legend=False)
            axes[i,j].set_ylabel(stadtteil)
            axes[i,j].autoscale(enable=True, axis='both', tight=None)
            
    handles, labels = axes[-1,-1].get_legend_handles_labels()
    fig.legend(handles, labels, loc='right')
    
    #set_size(10,10)
    plt.show()
    fig.savefig('./output/predict_parties_test.png', format='png')

    
plotAll([df_ypredGradientBoosting, df_ypredRandomForestRegressor]
        ,['GradientBoosting','RandomForestRegressor'])
```
